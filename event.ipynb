```python
import numpy as np # linear algebra
import pandas as pd 
note = pd.read_excel('event.xlsx')
note.head()
```

```python
note.isnull().sum()
```

```python
note.groupby('event_name')['note'].mean().sort_values(ascending=False).head() 
```

```python
note.groupby('event_name')['note'].count().sort_values(ascending=False).head() 
```

```python
ratings = pd.DataFrame(note.groupby('event_name')['note'].mean())
ratings['num of ratings'] =pd.DataFrame(note.groupby('event_name')['note'].count())
ratings.head()
```

```python
#Sort events based on note
note = note.sort_values('note', ascending=False)
note[['event_name', 'note']]
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.barh(note['event_name'],note['note'], align='center',color='skyblue')
plt.gca().invert_yaxis()
plt.xlabel("note")
plt.title("Popular events")
```
