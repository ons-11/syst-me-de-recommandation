```python
#Import TfIdfVectorizer from scikit-learn
from sklearn.feature_extraction.text import TfidfVectorizer

#Define a TF-IDF Vectorizer Object. Remove all english stop words such as 'the', 'a'
tfidf = TfidfVectorizer(stop_words='english')

#Replace NaN with an empty string
data['event_type'] = data['event_type'].fillna('')

#Construct the required TF-IDF matrix by fitting and transforming the data
tfidf_matrix = tfidf.fit_transform(data['event_type'])

#Output the shape of tfidf_matrix
tfidf_matrix.shape
```

```python
# Import linear_kernel
from sklearn.metrics.pairwise import linear_kernel
# Compute the cosine similarity matrix
cosine_sim = linear_kernel(tfidf_matrix,tfidf_matrix)
```

```python
cosine_sim
```

```python
#Construct a reverse map of indices and movie titles
indices = pd.Series(data.index, index=data['event_name']).drop_duplicates()
```

```python
indices[:10]
```

```python
# Function that takes in event title as input and outputs most similar eventsbased on type 
def get_recommendations(event_name,  cosine_sim=cosine_sim):
                        
    # Get the index of the movie that matches the event_name
    idx = indices[event_name]
    
     # Get the pairwsie similarities of all events with that event
    similarity = list(enumerate(cosine_sim[idx]))

    # Sort the events based on the similarity 
    similarity = sorted(similarity,key=lambda x: x[1], reverse=True)

    # Get the names of the 10 most similar events
    similarity = similarity[1:11]

    # Get the event indices
    event_indices = [i[0] for i in similarity]

    return data[['event_name','event_type']].iloc[event_indices].head(10)
```

```python
#test of the function 
get_recommendations('Rose Parade')
```

```python
#test of the function 
get_recommendations('Australia Week 2014 / G’Day USA')
```
